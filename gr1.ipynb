```python
def grGetTestRecords(filename, event_name, instrument, old_columns, new_columns):
    if filename.lower().endswith("csv"):
        df = pd.read_csv(filename, dtype=str) 
    elif filename.lower().endswith("xlsx"):
        df = pd.read_excel(filename, dtype=str)
    else:
        raise RuntimeError("Unknown filetype: " + filename)
        
    dict = {old_columns[i]: new_columns[i] for i in range(len(old_columns))}
    df2 = df[old_columns]
    df3 = df2.rename(columns=dict)
    df4 = df3[~df3['test_record_num'].isna()].copy()
    
    # get FXS id
    last = next(filter(lambda id: not pd.isna(id), df['Last Name']), None)
    if last == None:
        last = 0
        print("WARNING: No FXS ID found.")
    
    # don't forget to add key column
    # TODO: HARD CODED KEY COLUMN NAME study_id
    df4.insert(0, "study_id", last)
    
    # and don't forget to add the 'redcap_repeat_instance' column and its friend 'redcap_event_name'
    # We use 'new' for the redcap_repeat_instance. The event name has to come with the folder name 
    # as input. 
    df4.insert(1, "redcap_event_name", event_name)
    df4.insert(2, "redcap_repeat_instrument", instrument)
    df4.insert(3, "redcap_repeat_instance", "new")
    
    # and insert the filename
    df4.insert(4, 'input_filename', os.path.basename(filename))
    
    # Assign testtype based on contents of 'comments'
    df4.insert(5, 'testtype', 5)
    df4.loc[df4.comments.str.contains("control", na=False), 'testtype'] = 1
    df4.loc[df4.comments.str.contains("light", na=False), 'testtype'] = 2
    df4.loc[df4.comments.str.contains("loaded", na=False), 'testtype'] = 3
    
    return df4
```
